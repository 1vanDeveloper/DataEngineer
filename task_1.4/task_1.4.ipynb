```python
import numpy as np
import matplotlib.pyplot as plt
import time
import math
from IPython import display 
```

## 1. Построение распределений

Сгенерировать выборку с распределением Рэлэя ( https://ru.wikipedia.org/wiki/Распределение_Рэлея ) используя numpy и scipy и построить на основе ее кривую плотности распределения на основе гистограммы из Matplotlib.нове ее кривую плотности распределения на основе гистограммы из Matplotlib.

```python
for moda in np.arange(0.5, 4.5, 0.5):
  x = np.arange(0, 10, 0.1)
  y = (x / moda**2)*np.exp(-(x**2)/(2*moda**2))
  plt.plot(x, y, label=f'moda={moda}')
  plt.legend()
  display.display(plt.gcf())
  display.clear_output(wait=True)
  time.sleep(1)
```

```python
# кривая плотности распределения на основе гистограммы
for moda in np.arange(0.5, 4.5, 0.5):
  x = np.arange(0, 10, 0.1)
  y = (x / moda**2)*np.exp(-(x**2)/(2*moda**2))
  plt.hist(y, label=f'moda={moda}')
  plt.legend()
  display.display(plt.gcf())
  display.clear_output(wait=True)
  time.sleep(1)
```

## 2. Построение графика по функции

Сгенерировать массивы по функции (x^2 + y^2 - 1)^3 - x^2 y^3 = 0 и построить на основе них график.

```python
x_h = []
y_h = []
for x in np.arange(-2, 2.5, 0.003):
  for y in np.arange(-1.5, 1.5, 0.003):
    if abs((x**2 + y**2 - 1)**3 - x**2 * y**3) < 0.001:   
        x_h.append(x)
        y_h.append(y)

x_h = np.array(x_h)
y_h = np.array(y_h)

plt.scatter(x_h, y_h, color = '#FF0000')
plt.show()

```

## 3. Анализ зависимостей

Найти зависимости представленные в датасете приложенном к заданию.

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
insurance = []

d_region = {
    "northeast": 0,
    "northwest": 1,
    "southeast": 2,
    "southwest": 3
}

with open("./dataset_home.txt", "r") as file:
  for human in file.readlines():
    splited = human.split(",")
    age.append(int(splited[0]))
    sex.append(1 if splited[1] == "male" else 2)
    bmi.append(float(splited[2]))
    children.append(int(splited[3]))
    smoker.append(1 if splited[4] == "no" else 2)
    region.append(d_region[splited[5]])
    insurance.append(float(splited[6]))

age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
insurance = np.array(insurance)
```

### 3.1 Отношение курящих и некурящих людей в зависимости от кол-ва детей

```python
plt.hist(children[smoker == 1], label="некурящие")
plt.hist(children[smoker == 2], label="курящие")

plt.legend()
plt.show()
```

### 3.2 Зависимость стоимости страховки от индекса массы тела с учетом курения 

```python
plt.scatter(bmi[smoker == 2], insurance[smoker == 2], label="курящие")
plt.scatter(bmi[smoker == 1], insurance[smoker == 1], label="некурящие")

plt.legend()
plt.show()
```

### 3.3 Зависимость стоимости страховки от возраста в зависимости от пола

```python
plt.bar(age[sex == 1], insurance[sex == 1], label="мужчины")
plt.bar(age[sex == 2], insurance[sex == 2], label="женщины")

plt.legend()
plt.show()
```
